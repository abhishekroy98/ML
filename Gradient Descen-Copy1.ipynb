```python
def sgGI(data,datay,lr,m):
    N,n=data.shape
#     ms=np.zeros(n)
#     nm=np.zeros(n)
#     ms=np.zeros(n)
#     shape=ms.shape
    for j in range(N):
        y=datay[j:j+1,0:1]
        x=data[j:j+1,:]
#         print("x.shape=",x.shape)
#         print("y.shape=",y.shape)
#         ms=np.zeros(n)
#         print(ms)
#         print(x.shape)
#         print("m.shape=",m.shape)
#         x=np.append(x,[1])
        ms=(-2/N)*(y-(m*x).sum())*x
#         print("ms.shape=",ms.shape)
        m=(m-(lr*(ms))).reshape(m.shape)
#         print("ms.shapeA=",ms.shape)
#         nm=np.reshape(nm,m.shape)
#         print("nm.shape=",nm.shape)
#         m=nm
    return m
```

```python
def cost(data,datay,m):
    tc=0;
#     print('m',m)
    N,n=data.shape
#     print(data.shape)
    x=data[0:1,:]
#     print(x.shape)
#     print('gfuykdk',m)
#     m=m.reshape(x.shape)
#     print(m.shape)
#     print(y.shape)
    for i in range(N):
        x=data[i:i+1,:]
#         print(x)
#         print("\n\n\n")
        y=datay[i:i+1,0:1]
        tc+=(y-(m*x).sum())**2
#         print(tc,tc,tc)
    return tc/N
```

```python
def gdG(data,lr,it):
    N,M=data.shape
    m=np.zeros(M);
    x=data[:,0:M-1]
    one=np.ones((N,1))
#     print(one)
#     print(x)
    x=np.append(x,one,axis=1)
#     print("amitamit",x)
    x=scaler.transform(x)
    y=data[:,M-1:M]
    for i in range (it):
        m=sgGI(x,y,lr,m) #single step
        print(i, " Cost: ",cost(x,y,m))
    return m
```

```python
def run():
    data=np.genfromtxt("boston_train.csv",delimiter=",")
    lr=0.9
    it=150
    m=gdG(data,lr,it)
    return m
```

```python
def predict(m,data):
    N,n=data.shape
    y=np.zeros(N)
    for i in range(N):
        x=data[i,0:n]
        x=np.append(x,[1])
        for j in range(n+1):
            y[i]+=m[j]*x[j]
    return y
        
```
