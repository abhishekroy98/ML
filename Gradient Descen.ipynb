```python
def cost(data,datay,m):
    tc=0;
    N,n=data.shape
    x=data[0:1,:]
    for i in range(N):
        x=data[i:i+1,:]
        y=datay[i:i+1,0:1]
        tc+=(y-(m*x).sum())**2
    return tc/N
```

```python
def gdG(data,lr,it):
    N,M=data.shape
    m=np.zeros(M);
    x=data[:,0:M-1]
    one=np.ones((N,1))
    x=np.append(x,one,axis=1)
#     x=scaler.transform(x)
    y=data[:,M-1:M]
    for i in range (it):
        m=sgGI(x,y,lr,m) #single step
        print(i, " Cost: ",cost(x,y,m))
    return m
```

```python
def run():
    data=np.genfromtxt("boston_train.csv",delimiter=",")
    lr=0.9
    it=1500
    m=gdG(data,lr,it)
    return m
```

```python
def predict(m,data):
    N,n=data.shape
    y=np.zeros(N)
    for i in range(N):
        x=data[i,0:n]
        x=np.append(x,[1])
        for j in range(n+1):
            y[i]+=m[j]*x[j]
    return y
        
```
