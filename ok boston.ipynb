```python
def sgG(data,lr,m):
    N,n=data.shape
    ms=np.zeros(n)
    for j in range(N):
        x=data.iloc[j:j+1,0:n-1]
        x=x.values
#         x=x.reshape(1,-1)
#         x=scaller.transform(x)
#         x.reshape(-1,1)
        x=np.append(x,[1])
        y=data.iloc[j:j+1,n-1:n]
        M=(-2/N)*(y-(m*x).sum()).values
        Z=(M*x)
        ms=Z
    nm = m-(lr*ms)
    return nm
```

```python
def cost(data,m):
    tc=0;
    N,n=data.shape
    for i in range(N):
        x=data.iloc[i:i+1,0:n-1]
        x=x.values
#         x=x.reshape(1,-1)
#         x=scaller.transform(x)
#         x.reshape(-1,1)
        x=np.append(x,[1])
        y=data.iloc[i:i+1,n-1:n].values
        tc+=(y-((m*x).sum()))**2
    return tc
```
