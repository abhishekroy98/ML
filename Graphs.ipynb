```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4]
y=[2,4,6,8]
plt.scatter(x,y)
plt.show()
```

```python
x=[1,2,3,4]
y=[2,4,6,8]
plt.plot(x,y)
plt.scatter(x,y)
plt.show()
```

```python
x=[1,2,3,4]
y1=[2,4,6,8]
y2=[3,6,9,12]
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

```python
# add color
# first char as color
# second is for plot
plt.plot(x,y2,'g+')
```
